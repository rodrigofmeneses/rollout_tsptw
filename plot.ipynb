```python
import matplotlib.pyplot as plt
import numpy as np
from dados import (
    dumas_20, dumas_40, dumas_60,
    gendreau_20, gendreau_40, gendreau_60
)
from math import ceil
```

```python
dumas = [20, 40, 60, 80]
gendreau = [120, 140, 160, 180, 200]
pos_dumas = np.linspace(0, 2, 4)
pos_gendreau = np.linspace(0, 2.5, 5)
max_dumas = ceil(
    max(max(instance) for instance in [dumas_20, dumas_40, dumas_60])
) + 3
max_gendreau = ceil(
    max(max(instance) for instance in [gendreau_20, gendreau_40, gendreau_60])
) + 20
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))
for i, data in enumerate((dumas_20, dumas_40, dumas_60)):
    axes[i].set_title('n = {}'.format(dumas[i]))
    axes[i].set_ylabel('w')
    axes[i].set_xlabel('Diferença Percentual')
    rects = axes[i].barh(pos_dumas, data, tick_label=dumas, height=0.4)
    axes[i].set_xlim(0, max_dumas)
    for rect in rects:
        x = rect.get_width() + 0.2
        y = rect.get_y() + 0.16
        if rect.get_width() == 0:
            axes[i].annotate(
            '', 
            xy=(x, y)
            )
        else:
            axes[i].annotate(
                '{:.2f}%'.format(rect.get_width()), 
                xy=(x, y)
            )
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))
for i, data in enumerate((gendreau_20, gendreau_40, gendreau_60)):
    axes[i].set_title('n = {}'.format(dumas[i]))
    axes[i].set_ylabel('w')
    axes[i].set_xlabel('Diferença Percentual')
    rects = axes[i].barh(pos_gendreau, data, tick_label=gendreau, height=0.4)
    axes[i].set_xlim(0, max_gendreau)
    for rect in rects:
        x = rect.get_width() + 2
        y = rect.get_y() + 0.16
        if rect.get_width() == 0:
            axes[i].annotate(
            '', 
            xy=(x, y)
            )
        else:
            axes[i].annotate(
                '{:.2f}%'.format(rect.get_width()), 
                xy=(x, y)
            )
```
